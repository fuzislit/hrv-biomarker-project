```
import os
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```
# folder with RR interval text files
rr_root = "/content/drive/MyDrive/RR Intervals"

# folder to save results
save_folder = "/content/drive/MyDrive/LSTM_Project"
os.makedirs(save_folder, exist_ok=True)

# model settings
seq_length = 50
step_size = 5
epochs = 20
batch_size = 128
gen_length = 50
max_signal_len = 1000

# use False for full run
DEBUG_MODE = False
DEBUG_SUBJECTS = 10
```

```

def load_rr_file(file_path):
    signal = np.loadtxt(file_path)
    signal = np.asarray(signal, dtype=np.float32)
    signal = signal[~np.isnan(signal)]
    return signal

rr_signals = {}

all_files = sorted([f for f in os.listdir(rr_root) if f.endswith(".txt")])

if DEBUG_MODE:
    all_files = all_files[:DEBUG_SUBJECTS]

for file_name in all_files:
    full_path = os.path.join(rr_root, file_name)
    rr_signals[file_name] = load_rr_file(full_path)

subject_names = list(rr_signals.keys())
rr_list = list(rr_signals.values())

print("Total subjects loaded:", len(rr_list))
print("Example subject names:", subject_names[:5])
```

```
all_min = min(np.min(sig) for sig in rr_list)
all_max = max(np.max(sig) for sig in rr_list)

print("Global min RR:", all_min)
print("Global max RR:", all_max)
```

```
plt.figure(figsize=(12, 5))

for i, name in enumerate(subject_names[:3]):
    plt.plot(rr_signals[name][:500], label=name)

plt.title("Real Patient RR Signals")
plt.xlabel("Beat Index")
plt.ylabel("RR Interval (ms)")
plt.legend()
plt.grid(True)
plt.show()
```

```
def create_sequences(signal, seq_length=50, step_size=5):
    X = []
    y = []

    if len(signal) <= seq_length:
        return X, y

    for i in range(0, len(signal) - seq_length, step_size):
        X.append(signal[i:i+seq_length])
        y.append(signal[i+seq_length])

    return X, y
```

```
def build_lstm_model(seq_length=50):
    model = Sequential([
        LSTM(64, input_shape=(seq_length, 1), return_sequences=True),
        LSTM(32),
        Dense(16, activation="relu"),
        Dense(1)
    ])

    model.compile(
        optimizer="adam",
        loss="mse"
    )

    return model
```

```
def generate_rr(model, seed_sequence_norm, mean, std, length=50):
    generated_norm = list(seed_sequence_norm)
    current_seq = np.array(seed_sequence_norm, dtype=np.float32)

    for _ in range(length):
        inp = current_seq.reshape(1, len(current_seq), 1)
        next_val_norm = model.predict(inp, verbose=0)[0][0]

        generated_norm.append(next_val_norm)
        current_seq = np.append(current_seq[1:], next_val_norm)

    generated_norm = np.array(generated_norm, dtype=np.float32)

    # convert back to milliseconds
    generated_ms = generated_norm * std + mean

    # patient-specific bounds, but keep inside dataset-wide physiological range
    lower = max(mean - 3 * std, 300)
    upper = min(mean + 3 * std, 2000)
    generated_ms = np.clip(generated_ms, lower, upper)

    return generated_ms
```

```
#old
def generate_rr(model, seed_sequence, length=200):
    generated = list(seed_sequence)

    current_seq = np.array(seed_sequence)

    for _ in range(length):
        inp = current_seq.reshape(1, len(current_seq), 1)
        next_val = model.predict(inp, verbose=0)[0][0]

        generated.append(next_val)

        # slide window forward
        current_seq = np.append(current_seq[1:], next_val)

    return np.array(generated)
```

```
def compute_hrv(rr):
    rr = np.array(rr, dtype=np.float32)
    diff_rr = np.diff(rr)

    mean_rr = np.mean(rr)
    sdnn = np.std(rr)
    rmssd = np.sqrt(np.mean(diff_rr**2)) if len(diff_rr) > 0 else np.nan

    return {
        "mean_rr": mean_rr,
        "sdnn": sdnn,
        "rmssd": rmssd
    }
```

```
def compute_rmse(true_rr, pred_rr):
    min_len = min(len(true_rr), len(pred_rr))
    return np.sqrt(np.mean((true_rr[:min_len] - pred_rr[:min_len])**2))
```

```
print("Starting LOOCV...")
print("Number of subjects:", len(rr_list))

rmse_list = []
all_results = []

for i in range(len(rr_list)):
    print(f"LOOCV iteration {i+1}/{len(rr_list)} - testing {subject_names[i]}")

    # left-out subject
    test_signal = rr_list[i][:max_signal_len]

    # remaining subjects for training
    train_signals = []
    for j in range(len(rr_list)):
        if j != i:
            train_signals.append(rr_list[j][:max_signal_len])

    # build training data from normalized training subjects
    X_all = []
    y_all = []

    for signal in train_signals:
        mean = np.mean(signal)
        std = np.std(signal)
        if std == 0:
            std = 1.0

        signal_norm = (signal - mean) / std

        X_seq, y_seq = create_sequences(
            signal_norm,
            seq_length=seq_length,
            step_size=step_size
        )

        X_all.extend(X_seq)
        y_all.extend(y_seq)

    X_all = np.array(X_all, dtype=np.float32)
    y_all = np.array(y_all, dtype=np.float32)
    X_all = X_all.reshape((X_all.shape[0], X_all.shape[1], 1))

    # build and train model
    model = build_lstm_model(seq_length=seq_length)
    model.fit(
        X_all, y_all,
        epochs=epochs,
        batch_size=batch_size,
        verbose=0
    )

    # normalize left-out subject for generation
    test_mean = np.mean(test_signal)
    test_std = np.std(test_signal)
    if test_std == 0:
        test_std = 1.0

    test_signal_norm = (test_signal - test_mean) / test_std
    seed_norm = test_signal_norm[:seq_length]

    # generate RR in milliseconds
    generated = generate_rr(
        model,
        seed_sequence_norm=seed_norm,
        mean=test_mean,
        std=test_std,
        length=gen_length
    )

    real_compare = test_signal[:len(generated)]

    # evaluate ONLY after the seed
    real_eval = real_compare[seq_length:]
    generated_eval = generated[seq_length:]

    rmse = compute_rmse(real_eval, generated_eval)
    rmse_list.append(rmse)

    real_hrv = compute_hrv(real_eval)
    gen_hrv = compute_hrv(generated_eval)

    all_results.append({
        "subject": subject_names[i],
        "rmse": rmse,
        "real_mean_rr": real_hrv["mean_rr"],
        "gen_mean_rr": gen_hrv["mean_rr"],
        "real_sdnn": real_hrv["sdnn"],
        "gen_sdnn": gen_hrv["sdnn"],
        "real_rmssd": real_hrv["rmssd"],
        "gen_rmssd": gen_hrv["rmssd"]
    })
```

```
avg_rmse = np.mean(rmse_list)

print("RMSE for all subjects:")
print(rmse_list)

print("\nAverage RMSE across all subjects:")
print(avg_rmse)
```

```
# this takes a long RR signal and turns it into small training chunks
# X = previous beats
# y = next beat
def create_sequences(signal, seq_length=50, step_size=5):
    X = []
    y = []

    if len(signal) <= seq_length:
        return X, y

    for i in range(0, len(signal) - seq_length, step_size):
        X.append(signal[i:i+seq_length])
        y.append(signal[i+seq_length])

    return X, y
```

```
# simple LSTM model for next-beat prediction
def build_lstm_model(seq_length=50):
    model = Sequential([
        LSTM(64, input_shape=(seq_length, 1), return_sequences=True),
        LSTM(32),
        Dense(16, activation="relu"),
        Dense(1)
    ])

    model.compile(
        optimizer="adam",
        loss="mse"
    )

    return model
```

```
# this starts with a real short seed sequence,
# then keeps predicting the next beat over and over
def generate_rr(model, seed_sequence_norm, mean, std, length=50):
    generated_norm = list(seed_sequence_norm)
    current_seq = np.array(seed_sequence_norm, dtype=np.float32)

    for _ in range(length):
        inp = current_seq.reshape(1, len(current_seq), 1)
        next_val_norm = model.predict(inp, verbose=0)[0][0]

        generated_norm.append(next_val_norm)
        current_seq = np.append(current_seq[1:], next_val_norm)

    generated_norm = np.array(generated_norm, dtype=np.float32)

    # convert back to milliseconds
    generated_ms = generated_norm * std + mean

    # patient-specific clipping
    lower = mean - 3 * std
    upper = mean + 3 * std
    generated_ms = np.clip(generated_ms, lower, upper)

    return generated_ms
```

```
# these are the main HRV metrics your mentor mentioned
def compute_hrv(rr):
    rr = np.array(rr, dtype=np.float32)
    diff_rr = np.diff(rr)

    mean_rr = np.mean(rr)
    sdnn = np.std(rr)
    rmssd = np.sqrt(np.mean(diff_rr**2)) if len(diff_rr) > 0 else np.nan

    return {
        "mean_rr": mean_rr,
        "sdnn": sdnn,
        "rmssd": rmssd
    }
```

```
# RMSE tells us how close the generated RR intervals are to the real ones
def compute_rmse(true_rr, pred_rr):
    min_len = min(len(true_rr), len(pred_rr))
    return np.sqrt(np.mean((true_rr[:min_len] - pred_rr[:min_len])**2))
```

```
example_index = 0

test_signal = rr_list[example_index][:MAX_SIGNAL_LEN]

train_signals = []
for j in range(len(rr_list)):
    if j != example_index:
        train_signals.append(rr_list[j][:MAX_SIGNAL_LEN])

X_all = []
y_all = []

for signal in train_signals:
    mean = np.mean(signal)
    std = np.std(signal)
    if std == 0:
        std = 1.0

    signal_norm = (signal - mean) / std

    X_seq, y_seq = create_sequences(
        signal_norm,
        seq_length=SEQ_LENGTH,
        step_size=STEP_SIZE
    )

    X_all.extend(X_seq)
    y_all.extend(y_seq)

X_all = np.array(X_all, dtype=np.float32)
y_all = np.array(y_all, dtype=np.float32)
X_all = X_all.reshape((X_all.shape[0], X_all.shape[1], 1))

model = build_lstm_model(seq_length=SEQ_LENGTH)
model.fit(X_all, y_all, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=0)

test_mean = np.mean(test_signal)
test_std = np.std(test_signal)
if test_std == 0:
    test_std = 1.0

test_signal_norm = (test_signal - test_mean) / test_std
seed_norm = test_signal_norm[:SEQ_LENGTH]

generated = generate_rr(
    model,
    seed_sequence_norm=seed_norm,
    mean=test_mean,
    std=test_std,
    total_length=SEQ_LENGTH + GEN_LENGTH,
    predict_chunk=PREDICT_CHUNK,
    global_min=all_min,
    global_max=all_max
)

real_compare = test_signal[:len(generated)]

real_seed = real_compare[:SEQ_LENGTH]
real_future = real_compare[SEQ_LENGTH:]
generated_future = generated[SEQ_LENGTH:]

plt.figure(figsize=(12, 5))
plt.plot(range(SEQ_LENGTH), real_seed, label="Seed (Real Input)", linestyle="--")
x_future = range(SEQ_LENGTH, SEQ_LENGTH + len(real_future))
plt.plot(x_future, real_future, label="Real Future RR")
plt.plot(x_future, generated_future, label="Generated Future RR")
plt.title(f"Real vs Generated RR Intervals: {subject_names[example_index]}")
plt.xlabel("Beat Index")
plt.ylabel("RR Interval (ms)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{save_folder}/real_vs_generated_example_updated.png")
plt.show()

print("Generated min:", np.min(generated))
print("Generated max:", np.max(generated))
print("First 10 generated:", generated[:10])
```
