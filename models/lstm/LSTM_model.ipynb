```
import os
import numpy as np
import matplotlib.pyplot as plt

rr_root = "/content/drive/MyDrive/RR Intervals"

def load_rr_file(file_path):
    signal = np.loadtxt(file_path)
    signal = np.asarray(signal, dtype=np.float32)
    signal = signal[~np.isnan(signal)]
    return signal

rr_signals = {}

all_files = [f for f in os.listdir(rr_root) if f.endswith(".txt")]

#only using 10 patients so run time doesn't get affected
for file_name in all_files[:10]:
    full_path = os.path.join(rr_root, file_name)
    rr_signals[file_name] = load_rr_file(full_path)

print("Loaded patients:", len(rr_signals))
print(list(rr_signals.keys())[:5])
```

```
plt.figure(figsize=(12,5))

for i, name in enumerate(list(rr_signals.keys())[:3]):
    plt.plot(rr_signals[name][:500], label=name)

# plotting first 3 patients
plt.title("Real Patient RR Signals")
plt.xlabel("Beat Index")
plt.ylabel("RR Interval")
plt.legend()
plt.grid(True)
plt.show()
```

```
def create_sequences(signal, seq_length=20, step_size=10):
    X = []
    y = []

# if signal too short, skip

    if len(signal) <= seq_length:
        return X, y

# move through signal in steps
    for i in range(0, len(signal) - seq_length, step_size):
        X.append(signal[i:i+seq_length])
        y.append(signal[i+seq_length])

    return X, y

X_all = []
y_all = []

seq_length = 20
step_size = 10

for file_name, signal in rr_signals.items():
    # normalizing so all patients are comparable
    mean = np.mean(signal)
    std = np.std(signal)

    if std == 0:
        std = 1.0

    # only using first 1000 beats for speed
    signal_norm = (signal - mean) / std
    signal_norm = signal_norm[:1000]

    # create sequences
    X_seq, y_seq = create_sequences(signal_norm, seq_length, step_size)

    X_all.extend(X_seq)
    y_all.extend(y_seq)

# converting to arrays
X_all = np.array(X_all, dtype=np.float32)
y_all = np.array(y_all, dtype=np.float32)

print("X shape:", X_all.shape)
print("y shape:", y_all.shape)
```

```
from sklearn.model_selection import train_test_split

# reshaping for LSTM (samples, time steps, features)
X_all = X_all.reshape((X_all.shape[0], X_all.shape[1], 1))

# split into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X_all, y_all, test_size=0.2, random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential([
    LSTM(64, input_shape=(seq_length, 1), return_sequences=True),
    Dropout(0.2), # helps prevent overfitting

    LSTM(32),
    Dropout(0.2),

    Dense(16, activation="relu"),  # output = next RR value
    Dense(1)
])

model.compile(
    optimizer="adam",
    loss="mse",       # error between prediction and real value
    metrics=["mae"]
)

model.summary()
```

```
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=5,
    batch_size=128,
    callbacks=[early_stop],
    verbose=1
)
```

```
plt.figure(figsize=(8,5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(history.history["mae"], label="Train MAE")
plt.plot(history.history["val_mae"], label="Validation MAE")
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.title("Model Error")
plt.legend()
plt.grid(True)
plt.show()
```

```
y_pred = model.predict(X_test).flatten()

plt.figure(figsize=(10,5))
plt.plot(y_test[:200], label="Actual")
plt.plot(y_pred[:200], label="Predicted")
plt.title("Actual vs Predicted RR")
plt.legend()
plt.grid(True)
plt.show()
```

```
sample_idx = 0

input_seq = X_test[sample_idx].flatten()
actual_next = y_test[sample_idx]
pred_next = y_pred[sample_idx]

plt.figure(figsize=(8,4))
plt.plot(range(seq_length), input_seq, marker='o', label="Input")
plt.scatter(seq_length, actual_next, label="Actual")
plt.scatter(seq_length, pred_next, label="Predicted")

plt.title("Next Beat Prediction")
plt.legend()
plt.grid(True)
plt.show()
```
